```python
arr1 = rng.integers(1,49,size=(3,6))
arr2 = rng.integers(1,49,size=(6,3))
arr1
```

```python
# compute the union of arr1 and arr2
np.union1d(arr1, arr2)
```

```python
# compute the dot product of arr1 and arr2
np.dot(arr1, arr2)

# assign arr3 as the matrix multiplication of arr1 and arr2
arr3 = np.matmul(arr1, arr2)
arr3
```

```python
# combine_first example, merging two DataFrames by filling in missing values from another DataFrame.
# if no missing values, the original DataFrame is returned unchanged.
from numpy import NaN


df1 = pd.DataFrame({
    'A': [1, 2, NaN],
    'B': [4, 5, 6]
})
df2 = pd.DataFrame({
    'A': [7, 8, 9],
    'B': [10, 11, 12]
})
df1.combine_first(df2)
```

```python
# what is the purpose of the isna and notna methods in pandas?
# The isna() method in pandas is used to detect missing values (NaN) in a DataFrame or Series, returning a boolean mask indicating the presence of NaN values. 
# Conversely, the notna() method returns a boolean mask indicating the presence of non-missing values

# what is the purpose of the na_position parameter in the pandas sort_values method?
# The na_position parameter in the pandas sort_values method specifies the position of NaN values in
# the sorted result. It can take two values: 'first' (default) places NaNs at the beginning, 
# while 'last' places them at the end of the sorted data.
df = pd.DataFrame({
    'A': [3, 1, np.nan, 2],
    'B': [np.nan, 2, 3, 1]
})
df.sort_values(by='A', na_position='last')  # NaNs at the end


```
