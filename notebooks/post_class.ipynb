```python
import numpy as np
import pandas as pd 
```

```python
# Create a sample DataFrame with mixed types
dft = pd.DataFrame({
    'A': np.random.randint(0, 100, 10),
    'B': np.random.rand(10),
    'C': np.random.randint(0, 100, 10).astype('uint'),
    'D': ['text'] * 10
})
```

```python
dft
```

```python
# Select all numeric columns except float from the DataFrame dft
dft.select_dtypes(include=['int', 'uint'])
```

```python
#return the last 3 rows of the DataFrame dft
dft.tail(3)
```

```python
# Return the minimum and maximum of a Series x as a new Series with the index ["min", "max"]
def min_max(x):
    return pd.Series([x.min(), x.max()], index=["min", "max"])
dft['A'].apply(min_max) 
```
